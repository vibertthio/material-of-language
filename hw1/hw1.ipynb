```python
import math
import random

emo_sea = '🌊'
emo_drip = '💧'
# emo_things = ['🐚', '🦑', '⛵', '🐙', '🏝️', '🐟', '🐠', '🐡', '🦈']
emo_things = ['⛵', '🏝️', '🐟', '🐠', '🐡', '🦈']

def getRandomEmojiOfThings():
    return emo_things[random.randint(0, len(emo_things) - 1)]

def drawOceanEcosystem(pure_water = True):
    width = 50
    freq = 0.8
    amp = 3
    layers_num = 3
    layer_dist = 6
    pos = {}
    y_max = 0
    for j in range(layers_num):
        for i in range(width):
            x = i
            y = math.floor(math.sin(i * 0.5) * amp) + amp * 2 + j * layer_dist
            y_max = max(y, y_max)
            if y in pos.keys():
                pos[y].append(x)
            else:
                pos[y] = [x]

    for y in range(y_max + 1):
        output = ''
        if y not in pos.keys():
            continue
        for x in range(width):
            if x in pos[y]:
                output += emo_sea
            else:
                if (pure_water or random.random() < 0.95):
                    output += emo_drip
                else:
                    output += getRandomEmojiOfThings()
        print(output)
```

```python
drawOceanEcosystem()
```

```python
symbol_flat = '_'
symbol_rock = '^'
width = 110
height = 50
x_ratio = 2
y_ratio = 1

def checkRockOrNot(x, y, point):
    center_x, center_y, wavelength, w_number = point
    dist = pow(pow((x - center_x) / x_ratio, 2) + pow((y - center_y) / y_ratio, 2), 0.5)
    isRock = math.floor(dist) % wavelength is 0 and dist < w_number * wavelength
    return isRock

def drawKaresansui(points):
    for y in range(height):
        output = ''
        for x in range(width):
            isRock = False
            for point in points:
                if checkRockOrNot(x, y, point):
                    isRock = True
                    break
            if isRock:
                output += symbol_rock
            else:
                output += symbol_flat
        print(output)
```

```python
points = [(30, 5, 4, 5), (50, 40, 8, 4), (100, 22, 4, 3), (110, 0, 4, 3)]
drawKaresansui(points)
```

```python
def drawMultipleKaresansui(num):
    points = []
    for i in range(num):
        points.append((
            random.randint(0, width),
            random.randint(0, height),
            random.randint(3, 8),
            random.randint(3, 5)
        ))
    drawKaresansui(points)

drawMultipleKaresansui(1)
print()
drawMultipleKaresansui(2)
print()
drawMultipleKaresansui(4)
```
